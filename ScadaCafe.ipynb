```python
# Função para calcular comissão individual:
def calcular_comissao(dadosdic: dict, mes_desejado: str):
  if mes_desejado.lower() not in dados_2025:
    print(f'O mês {mes_desejado} não foi encontrado')
    return []
  registros = dados_2025[mes_desejado.lower()]
  resultados = []
  total = 0

  for dia in registros:
    comissao_total_dia = dia["Comissao"]
    equipe_dia = dia["EquipeDia"]
    comissao_individual_dia = comissao_total_dia * 0.80 / equipe_dia
    total += comissao_individual_dia
    #print(f"Data: {dia['Data']} - Comissão: {comissao_individual_dia:.2f}")

    resultados.append({
      "Mes": mes_desejado,
      "Data": dia['Data'],
      "Comissao_Individual": round(comissao_individual_dia, 2)
    })
  #print(f"Comissão acumulada {total:.2f}")

  return resultados

#print(calcular_comissao(dados_2025, 'outubro'))
```

```python
# Função para calcular Ticket Médio Dia:
def calcular_ticket(dadosdic: dict, mes_desejado: str):
  if mes_desejado.lower() not in dados_2025:
    print(f'O mês {mes_desejado} não foi encontrado')
    return []
  registros = dados_2025[mes_desejado.lower()]
  total_cupons_dia = 0
  total_fat_dia = 0

  for dia in registros:
    fat_dia = dia["Faturamento"]
    cupons_dia = dia["Cupons"]
    ticket_medio_dia = fat_dia / cupons_dia
    total_cupons_dia += cupons_dia
    total_fat_dia += fat_dia
    print(f"Data: {dia['Data']} - Ticket: {ticket_medio_dia:.2f}")

  ticket_medio_mes = total_fat_dia / total_cupons_dia

  return total_cupons_dia, total_fat_dia, ticket_medio_mes

#print(calcular_ticket(dados_2025, 'outubro'))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Selecionar o ano desejado
ano = 2025
dados_ano = dados_acum[ano]

# Extrair listas
meses = [item["Mês"].capitalize() for item in dados_ano]
metas = [item["Meta"] for item in dados_ano]
faturamentos = [item["Faturamento"] for item in dados_ano]

# Calcular percentual de atingimento
percentuais = [(f / m * 100) if m != 0 else 0 for f, m in zip(faturamentos, metas)]

# Criar gráfico
x = np.arange(len(meses))
largura = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - largura/2, faturamentos, largura, label="Faturamento", color="#1f77b4")
plt.bar(x + largura/2, metas, largura, label="Meta", color="#ff7f0e")

# Personalização
plt.title(f"Faturamento x Meta - {ano}", fontsize=16, fontweight="bold")
plt.xlabel("Mês", fontsize=10)
plt.ylabel("Valor (R$)", fontsize=10)
plt.xticks(x, meses, rotation=45)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Exibir valores e percentuais acima das barras
for i, (fat, meta, perc) in enumerate(zip(faturamentos, metas, percentuais)):
    plt.text(i - largura, fat + (meta * 0.01), f"R$ {fat:,.0f}", ha="center", fontsize=9)
    plt.text(i + largura, meta + (meta * 0.01), f"R$ {meta:,.0f}", ha="center", fontsize=9, color="#ff7f0e")
    if fat > 0:
        plt.text(i, max(fat, meta) + (meta * 0.03), f"{perc:.1f}%", ha="center", fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```
