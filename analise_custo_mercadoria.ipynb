## Importação de bases e tratamento de dados:

```python
import pandas as pd

arquivo_bps = 'export-BPS.csv'
arquivo_cinema = 'export-CINEMA.csv'

df_bps = pd.read_csv(arquivo_bps)
df_cinema = pd.read_csv(arquivo_cinema)

display(df_bps.head())
display(df_cinema.head())
```

```python
# Agrupamento das bases:
comparativo_custos = pd.merge(
    df_bps[['Código', 'Item de Estoque', 'Descrição Grupo', 'Unidade', 'Custo Médio']],
    df_cinema[['Código', 'Custo Médio']],
    on='Código',
    suffixes=('_BPS', '_CINEMA'),
    how='outer'
)

# Corrigir pontuação das colunas de Custo, alterando a "," por ".":
comparativo_custos["Custo Médio_BPS"] = comparativo_custos["Custo Médio_BPS"].str.replace(",", ".")
comparativo_custos["Custo Médio_CINEMA"] = comparativo_custos["Custo Médio_CINEMA"].str.replace(",", ".")
# Transformar colunas de Custo Médio em numéricas:
comparativo_custos['Custo Médio_BPS'] = pd.to_numeric(comparativo_custos['Custo Médio_BPS'], errors='coerce')
comparativo_custos['Custo Médio_CINEMA'] = pd.to_numeric(comparativo_custos['Custo Médio_CINEMA'], errors='coerce')

display(comparativo_custos)
comparativo_custos.info()
```

## Desenvolvimento e Análises:

```python
# Inserir coluna comparativa de custos:
comparativo_custos['Diferença de Custo'] = comparativo_custos['Custo Médio_CINEMA'] - comparativo_custos['Custo Médio_BPS']
display(comparativo_custos)
```

```python
# Inserir coluna de Status de Comparação:
import numpy as np

comparativo_custos["Status"] = np.where(
    comparativo_custos["Diferença de Custo"] > 0, "MAIOR",
    np.where(comparativo_custos["Diferença de Custo"] < 0, "MENOR", "-")
)

display(comparativo_custos)
```
